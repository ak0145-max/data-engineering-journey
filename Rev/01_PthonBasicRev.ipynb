## 🔤 TOPIC 3: String Operations

| Method | What it does | Example |
|--------|--------------|---------|
| `.strip()` | Remove spaces from both ends | `"  hi  ".strip()` → `"hi"` |
| `.lower()` | All lowercase | `"HI".lower()` → `"hi"` |
| `.upper()` | All uppercase | `"hi".upper()` → `"HI"` |
| `.title()` | Capitalize Each Word | `"hi there".title()` → `"Hi There"` |
| `.capitalize()` | Capitalize first letter only | `"hi there".capitalize()` → `"Hi there"` |
| `.replace(a, b)` | Replace text | `"cat".replace("c","b")` → `"bat"` |
| `.split(sep)` | Split into list | `"a,b".split(",")` → `["a","b"]` |

### 🔑 Chain methods:
```python
"  HELLO  ".strip().lower()   # → "hello"
```

````python
# ✅ SOLUTION — String Operations

raw_name  = "   rahim UDDIN   "
raw_email = "RAHIM@GMAIL.COM"
raw_city  = "dhaka"

clean_name  = raw_name.strip().title()   # strip spaces then title case
clean_email = raw_email.lower()          # lowercase
clean_city  = raw_city.capitalize()      # first letter capital

print(f"Name:  {clean_name}")
print(f"Email: {clean_email}")
print(f"City:  {clean_city}")

# f-string summary
summary = f"{clean_name} ({clean_email}) from {clean_city}"
print(summary)
````

## 🔄 TOPIC 4: Type Conversion

| Function | Converts To | Example |
|----------|-------------|----------|
| `int(x)` | Integer | `int("8")` → `8` |
| `float(x)` | Float | `float("1250.75")` → `1250.75` |
| `str(x)` | String | `str(42)` → `"42"` |
| `bool(x)` | Boolean | `bool(0)` → `False` |

### ⚠️ Common Mistakes:
```python
int("1250.75")   # ❌ ERROR! Can't convert decimal string to int
int(float("1250.75"))  # ✅ Convert to float first, then int
```

````python
# ✅ SOLUTION — Type Conversion

str_price    = "1250.75"
str_quantity = "8"
str_discount = "15"

price    = float(str_price)      # "1250.75" → 1250.75
quantity = int(str_quantity)     # "8"  → 8
discount = int(str_discount)     # "15" → 15

subtotal        = price * quantity
discount_amount = subtotal * discount / 100
final_total     = subtotal - discount_amount

print(f"Subtotal:         {subtotal:.2f} BDT")
print(f"Discount ({discount}%):  -{discount_amount:.2f} BDT")
print(f"Final Total:      {final_total:.2f} BDT")
````

## ✅ TOPIC 5: Comparisons & Booleans

| Operator | Meaning | Example |
|----------|---------|----------|
| `==` | Equal to | `5 == 5` → `True` |
| `!=` | Not equal | `5 != 3` → `True` |
| `>` | Greater than | `5 > 3` → `True` |
| `<` | Less than | `5 < 3` → `False` |
| `>=` | Greater or equal | `5 >= 5` → `True` |
| `<=` | Less or equal | `3 <= 5` → `True` |

### 🔗 Combining Booleans:
| Keyword | Meaning | Example |
|---------|---------|----------|
| `and` | Both must be True | `True and False` → `False` |
| `or` | At least one True | `True or False` → `True` |
| `not` | Flip it | `not True` → `False` |

````python
# ✅ SOLUTION — Comparisons & Booleans

rows_loaded       = 9500
rows_expected     = 10000
null_percentage   = 3.2
max_allowed_nulls = 5.0
run_time_minutes  = 12
max_run_time      = 15

data_complete     = rows_loaded >= rows_expected          # False (9500 < 10000)
data_clean        = null_percentage <= max_allowed_nulls  # True (3.2 <= 5.0)
within_time       = run_time_minutes <= max_run_time      # True (12 <= 15)

all_checks_passed = data_complete and data_clean and within_time
any_issue         = not all_checks_passed

print(f"Data complete:    {data_complete}")
print(f"Data clean:       {data_clean}")
print(f"Within time:      {within_time}")
print(f"All passed:       {all_checks_passed}")
print(f"Any issue:        {any_issue}")
````

## 🔀 TOPIC 6: Conditionals (if / elif / else)

```python
# Structure:
if condition1:
    # runs if condition1 is True
elif condition2:
    # runs if condition1 False, condition2 True
else:
    # runs if ALL above are False
```

### 🔑 Key Rules:
- Conditions are checked **top to bottom**
- Once one matches, the rest are **skipped**
- `else` has **no condition** — it's the default
- Indentation (4 spaces) is **mandatory** in Python!

````python
# ✅ SOLUTION — Conditionals

pipeline_score = 87   # 🔁 Change this value to test all branches!

if pipeline_score >= 90:
    grade = "EXCELLENT"
    color = "green"
elif pipeline_score >= 70:
    grade = "GOOD"
    color = "yellow"
elif pipeline_score >= 50:
    grade = "WARNING"
    color = "orange"
else:
    grade = "CRITICAL"
    color = "red"

print(f"Pipeline Score: {pipeline_score} -> {grade} ({color})")
````

## 🔁 TOPIC 7: For Loops

```python
# Loop over a list:
for item in my_list:
    print(item)

# Loop with index using range:
for i in range(len(my_list)):
    print(i, my_list[i])

# Loop list + another list together:
for name, count in zip(names, counts):
    print(name, count)
```

### 🔑 `zip()` is perfect for pairing two lists!

````python
# ✅ SOLUTION — For Loops

tables     = ["customers", "orders", "products", "reviews", "inventory"]
row_counts = [15000, 42000, 3500, 28000, 7200]

# 7a: Print each table in UPPERCASE
print("--- 7a: Uppercase table names ---")
for table in tables:
    print(table.upper())

print("---")

# 7b: Table name + row count (using range)
print("--- 7b: Table with row count ---")
for i in range(len(tables)):
    print(f"Table: {tables[i]} -> {row_counts[i]} rows")

print("---")

# 7c: Total rows
print("--- 7c: Total rows ---")
total = 0
for count in row_counts:
    total += count   # same as: total = total + count

print(f"Total rows across all tables: {total}")
````

````python
# 💡 BONUS: Same thing using zip() — cleaner!

print("--- Using zip() ---")
for table, count in zip(tables, row_counts):
    print(f"Table: {table} -> {count:,} rows")

# 💡 BONUS: Built-in sum()
print(f"\nTotal using sum(): {sum(row_counts):,}")
````

## 🔂 TOPIC 8: While Loop

```python
# Structure:
while condition:
    # runs as long as condition is True
    # ⚠️ MUST change something to avoid infinite loop!
```

### 🔑 Key Difference from For Loop:
| For Loop | While Loop |
|----------|------------|
| Known number of steps | Unknown number of steps |
| Loop over a list/range | Loop until a condition is met |
| `for item in list` | `while remaining > 0` |

````python
# ✅ SOLUTION — While Loop

remaining    = 7500
batch_size   = 1000
batch_number = 0

while remaining > 0:
    batch_number += 1
    if remaining >= batch_size:
        processed = batch_size
    else:
        processed = remaining     # last batch — process whatever's left
    remaining -= processed        # same as: remaining = remaining - processed
    print(f"Batch {batch_number}: processed {processed}, remaining: {remaining}")

print(f"Done! Processed in {batch_number} batches")
````

## 🏁 TOPIC 9: Putting It All Together — Pipeline Report

````python
# ✅ SOLUTION — Complete Pipeline Report

pipeline          = "ecommerce_daily_sync"
tables_processed  = ["users", "orders", "products", "payments"]
total_rows        = 95400
failed_rows       = 1230
start_time_seconds = 0
end_time_seconds   = 324

# Calculate metrics
success_rows    = total_rows - failed_rows
success_rate    = (success_rows / total_rows) * 100
duration_minutes = (end_time_seconds - start_time_seconds) / 60
rows_per_second = total_rows / (end_time_seconds - start_time_seconds)

# Determine status
if success_rate >= 98:
    status = "HEALTHY"
elif success_rate >= 95:
    status = "WARNING"
else:
    status = "CRITICAL"

# Print the report
print("=" * 50)
print(f"  PIPELINE REPORT: {pipeline}")
print("=" * 50)
print(f"  Status:     {status}")
print(f"  Duration:   {duration_minutes:.1f} minutes")
print(f"  Tables processed:")
for table in tables_processed:
    print(f"    - {table}")
print(f"  Total rows:  {total_rows:,}")
print(f"  Successful:  {success_rows:,} ({success_rate:.1f}%)")
print(f"  Failed:      {failed_rows:,}")
print(f"  Throughput:  {rows_per_second:.0f} rows/sec")
print("=" * 50)
````
