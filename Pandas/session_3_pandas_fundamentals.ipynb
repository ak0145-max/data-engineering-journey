## 2. Series — The Building Block

A **Series** is a single column of data — like one column in a spreadsheet.

```python
# Creating a Series from a list
prices = pd.Series([29.99, 49.99, 19.99])
print(prices)
print()
print(f"Type: {type(prices)}")
```

```python
# Series with custom index
prices = pd.Series(
    [29.99, 49.99, 19.99],
    index=["shirt", "pants", "cap"]
)
print(prices)
```

```python
# Key attributes
print(f"Values: {prices.values}")
print(f"Index:  {prices.index.tolist()}")
print(f"Dtype:  {prices.dtype}")
print(f"Size:   {prices.size}")
```

## 3. DataFrame — The Star of the Show

A **DataFrame** is a 2D table — like a full spreadsheet or database table.  
Think of it as **multiple Series** sharing the same index.

```python
# Creating a DataFrame from a dictionary
data = {
    "product": ["Laptop", "Phone", "Tablet"],
    "price":   [999.99, 699.99, 399.99],
    "stock":   [50, 150, 80]
}

df = pd.DataFrame(data)
print(df)
```

```python
# DataFrame with custom index
df = pd.DataFrame(data, index=["P001", "P002", "P003"])
print(df)
```

## 7. Selecting Columns

```python
# Recreate our orders DataFrame
orders = {
    "order_id":  [101, 102, 103, 104, 105, 106, 107, 108],
    "customer":  ["Alice", "Bob", "Alice", "Charlie", "Bob", "Dave", "Alice", "Charlie"],
    "product":   ["Laptop", "Phone", "Tablet", "Laptop", "Phone", "Mouse", "Keyboard", "Monitor"],
    "amount":    [999.99, 699.99, 399.99, 999.99, 699.99, 29.99, 79.99, 549.99],
    "city":      ["Dhaka", "Chittagong", "Dhaka", "Sylhet", "Dhaka", "Chittagong", "Sylhet", "Dhaka"],
}
df = pd.DataFrame(orders)

# Single column → returns a Series
print("=== Single Column (Series) ===")
print(df["product"])
```

```python
# Multiple columns → returns a DataFrame
print("=== Multiple Columns (DataFrame) ===")
print(df[["customer", "amount"]])
```

```python
# Using loc — select by label
print("=== loc: rows 0-2, specific columns ===")
print(df.loc[0:2, ["product", "amount"]])
```

```python
# Using iloc — select by position
print("=== iloc: first 3 rows, columns 1-3 ===")
print(df.iloc[0:3, 0:3])
```

## 8. Filtering Rows

Ask questions of your data — Pandas gives you the answers.

```python
# Single condition: orders over ৳500
expensive = df[df["amount"] > 500]
print("=== Orders over ৳500 ===")
print(expensive)
```

```python
# Multiple conditions with AND (&)
# Alice's orders over ৳500
alice_expensive = df[(df["amount"] > 500) & (df["customer"] == "Charlie")]
print("=== Alice's orders over ৳500 ===")
print(alice_expensive)
```

```python
# Multiple conditions with OR (|)
# Laptops or Phones
devices = df[(df["product"] == "Laptop") | (df["product"] == "Phone") ]
print("=== Laptops or Phones ===")
print(devices)
```

```python
# Using isin() for multiple values — cleaner than multiple OR
devices = df[df["product"].isin(["Laptop", "Phone", "Tablet"])]
print("=== Devices (using isin) ===")
print(devices)
```

```python
# Filter by city
dhaka_orders = df[df["city"] == "Dhaka"]
print(f"=== Orders from Dhaka ({len(dhaka_orders)} orders) ===")
print(dhaka_orders)
```

## 9. Sorting Data

```python
# Sort by amount — ascending (default)
print("=== Sorted by amount (ascending) ===")
print(df.sort_values("amount"))
```

```python
# Sort by amount — descending
print("=== Sorted by amount (descending) ===")
print(df.sort_values("amount", ascending=False))
```

```python
# Sort by multiple columns
print("=== Sorted by customer, then amount (desc) ===")
print(df.sort_values(["amount", "customer"], ascending=[True, False]))
```

```python
# Sort and reset index (clean output)
df_sorted = df.sort_values("amount", ascending=True).reset_index(drop=True)
print("=== Sorted with clean index ===")
print(df_sorted)
```

## 10. Adding & Modifying Columns

```python
print(df)
```

```python
# Add a new column — tax (5%)
df["tax"] = df["amount"] * 0.05
print("=== Added tax column ===")
print(df)
print(df[["product", "amount", "tax"]].head())
```

```python
# Column from calculation
df["total"] = df["amount"] + df["tax"]
print("=== Added total column ===")
print(df[["product", "amount", "tax", "total"]].head())
```

```python
# Conditional column using np.where
df["tier"] = np.where(df["amount"] > 500, "Premium", "Standard")
print("=== Added tier column ===")
print(df[["product", "amount", "tier"]])
```

```python
df["tier"] = "Standard"
print(df["tier"])
df.loc[df["amount"] > 500, "tier"] = "Premium"
print(df)
```

```python
# Rename columns
print(df.columns.tolist())
df = df.rename(columns={
    "amount": "order_amount",
    "customer": "customer_name"
})
print("=== Renamed columns ===")
print(df.columns.tolist())
```

```python
# Drop a column
df = df.drop(columns=["tax"])
print("=== Dropped tax column ===")
print(df.columns.tolist())
```

## 11. Basic Aggregations

```python
# Simple aggregation methods
print(df['order_amount'])
print(df['order_amount'].sort_values())
print(f"Total revenue: ৳{df['order_amount'].sum():,.2f}")
print(f"Average order: ৳{df['order_amount'].mean():,.2f}")
print(f"Median order:  ৳{df['order_amount'].median():,.2f}")
print(f"Min order:     ৳{df['order_amount'].min():,.2f}")
print(f"Max order:     ৳{df['order_amount'].max():,.2f}")
print(f"Unique products: {df['product'].nunique()}")
```

```python
# value_counts() — instant frequency table
print("=== Product frequency ===")
print(df["product"].value_counts())
print()
print("=== Customer frequency ===")
print(df["customer_name"].value_counts())
```

```python
# GroupBy — total per customer
print("=== Total spending per customer ===")
print(df.groupby(["customer_name"])["order_amount"].sum())
print(df.groupby(["customer_name"])["order_amount"].mean())
print(df.groupby(["customer_name"])["order_amount"].min())
```

```python
# GroupBy — multiple aggregations
print("=== Customer summary ===")
customer_summary = df.groupby("customer_name")["order_amount"].agg(
    ["sum", "mean", "count", "min", "max"]
)
print(customer_summary)
```

```python
# GroupBy multiple columns with custom columns
print("=== Revenue by city ===")
city_summary = df.groupby("city")["order_amount"].agg(["sum", "count"])
city_summary.columns = ["total_revenue", "num_orders"]
print(city_summary)
```

## 12. Hands-On Exercise: E-Commerce Analysis

Put it all together! Complete these tasks:

1. Filter orders where amount > ৳500
2. Add a "discount" column: 10% of order_amount
3. Add a "final_price" column: order_amount - discount
4. Sort by final_price descending
5. Group by customer — find total and average spending
6. Export the result

```python
# 🎯 YOUR TURN — Complete the exercise!

# Step 1: Filter expensive orders
expensive_orders = df[df["order_amount"] > 500]
print(f"Expensive orders: {len(expensive_orders)}")
print(expensive_orders[["customer_name", "product", "order_amount"]])
```

```python
# Step 2: Add discount column (10%)
expensive_orders = expensive_orders.copy()  # Avoid SettingWithCopyWarning
expensive_orders["discount"] = expensive_orders["order_amount"] * 0.10
print(expensive_orders[["product", "order_amount", "discount"]].head())
```

```python
# Step 3: Add final_price column
expensive_orders["final_price"] = expensive_orders["order_amount"] - expensive_orders["discount"]
print(expensive_orders[["product", "order_amount", "discount", "final_price"]].head())
```

```python
# Step 4: Sort by final_price descending
expensive_sorted = expensive_orders.sort_values("final_price", ascending=False).reset_index(drop=True)
print(expensive_sorted[["customer_name", "product", "final_price"]])
```

```python
# Step 5: Group by customer — total and average spending
customer_analysis = expensive_orders.groupby("customer_name")["final_price"].agg(
    ["sum", "mean", "count"]
)
customer_analysis.columns = ["total_spent", "avg_order", "num_orders"]
print("=== Customer Analysis ===")
print(customer_analysis)
```

```python
# Step 6: Export
expensive_sorted.to_csv("analysis_output.csv", index=False)
print("✅ Exported to analysis_output.csv!")
```

## 13. Best Practices Recap

✅ **Always explore first** — `head()`, `shape`, `info()`  
✅ **Use bracket notation** — `df["col"]` not `df.col`  
✅ **Chain operations** — `df.filter().sort_values().head()`  
✅ **Avoid loops** — use vectorized operations  
✅ **Save intermediate results** — export at each pipeline stage  
✅ **Name columns clearly** — `snake_case`, lowercase, no spaces
