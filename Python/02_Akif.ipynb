```python
# TASK 5: Functions 
#5a – average function
def cal_avg(numbers):
    return sum(numbers) / len(numbers)

print(cal_avg([80, 92, 75, 88, 95]))

#5b – grade function 
def get_gd(score=0):
    if   score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 60: return "D"
    else:             return "F"

print(get_gd(92))   
print(get_gd(75))   
print(get_gd(45))   

#5c – returning multiple values as a tuple
def summarize(scores):
    count = len(scores)
    avg   = sum(scores) / count
    return count, avg       

count, avg = summarize([85, 90, 78, 92])
print(f"Count: {count}, Average: {avg}")
```

```python
# 💡 SOLUTION — Task 6
import csv
import json

students = [
    {'name': 'Rahim',  'dept': 'CSE', 'cgpa': 3.65},
    {'name': 'Fatima', 'dept': 'EEE', 'cgpa': 3.82},
    {'name': 'Karim',  'dept': 'CSE', 'cgpa': 3.21},
    {'name': 'Nadia',  'dept': 'BBA', 'cgpa': 3.90},
    {'name': 'Arif',   'dept': 'EEE', 'cgpa': 2.95},
]

# 6a. Write CSV
with open('students.csv', 'w', newline='') as csvfile:
    fieldnames = ['name', 'dept', 'cgpa']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(students)

# 6b. Read CSV
with open('students.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        print(f"Name: {row['name']}, Dept: {row['dept']}, CGPA: {row['cgpa']}")

# 6c. Stats → JSON
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

cgpa_list = [s['cgpa'] for s in students]
summary = {
    'total_students': len(students),
    'average_cgpa':   round(calculate_average(cgpa_list), 2),
    'highest_cgpa':   max(cgpa_list),
    'lowest_cgpa':    min(cgpa_list)
}

with open('summary.json', 'w') as f:
    json.dump(summary, f, indent=2)

with open('summary.json') as f:
    result = json.load(f)
print(result)

print('\nHomework Done!')
```
